# Customer segmentation — exploration

Quick exploratory views of the synthetic customer dataset and preprocessing outputs.
Run `python main.py` from the project root for the full clustering pipeline and saved charts.

```python
import sys
from pathlib import Path

ROOT = Path.cwd().resolve()
if not (ROOT / "src").is_dir():
    ROOT = ROOT.parent
if str(ROOT) not in sys.path:
    sys.path.insert(0, str(ROOT))

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from IPython.display import display

from src.data_loader import load_customers
from src.preprocessor import fit_transform, NUMERIC_FEATURES

RANDOM_STATE = 42
np.random.seed(RANDOM_STATE)

df = load_customers(random_state=RANDOM_STATE)
if "true_segment" in df.columns:
    display(df.drop(columns=["true_segment"]).head())
else:
    display(df.head())

print(df.shape)
print(df.dtypes)
```

```python
sns.set_theme(style="whitegrid")
fig, axes = plt.subplots(2, 2, figsize=(11, 8))
sns.histplot(df["age"], kde=True, ax=axes[0, 0])
axes[0, 0].set_title("Age")
sns.histplot(df["income"], kde=True, ax=axes[0, 1])
axes[0, 1].set_title("Income")
sns.scatterplot(
    data=df,
    x="purchase_frequency",
    y="avg_order_value",
    hue="gender" if "gender" in df.columns else None,
    alpha=0.35,
    ax=axes[1, 0],
)
axes[1, 0].set_title("Frequency vs AOV")
sns.scatterplot(
    data=df,
    x="days_since_last_purchase",
    y="email_open_rate",
    alpha=0.35,
    ax=axes[1, 1],
)
axes[1, 1].set_title("Recency vs email open rate")
plt.tight_layout()
plt.show()
```

```python
pre = fit_transform(df, random_state=RANDOM_STATE)
print("Model matrix shape:", pre.X_model.shape)
print("PCA explained variance ratio:", pre.pca_2d.explained_variance_ratio_)

corr = pre.raw_frame[NUMERIC_FEATURES].corr()
plt.figure(figsize=(9, 7))
sns.heatmap(corr, annot=False, cmap="vlag", center=0)
plt.title("Numeric feature correlations (cleaned raw features)")
plt.show()
```
